````
# elde edilen aÄŸÄ±rlÄ±ÄŸÄ±n EMA sÄ±z optimize edilerek orjinal v5s e dÃ¶nÃ¼ÅŸÃ¼m iÅŸlemleri

import torch
import os

def strip_optimizer(path, save_path=None):
    """
    PyTorch 2.6+ uyumlu strip_optimizer fonksiyonu.
    `weights_only=False` parametresi ile checkpoint'i aÃ§ar,
    optimizer vb. kÄ±sÄ±mlarÄ± temizler, model aÄŸÄ±rlÄ±klarÄ±nÄ± FP16 yapar,
    ve kaydeder.
    """
    x = torch.load(path, map_location=torch.device('cpu'), weights_only=False)

    # EÄŸer EMA varsa model olarak onu kullan
    if x.get('ema'):
        x['model'] = x['ema']

    # optimizer, ema, vb. alanlarÄ± temizle
    for k in ['optimizer', 'best_fitness', 'wandb_id', 'ema', 'updates']:
        if k in x:
            x[k] = None

    x['epoch'] = -1
    x['model'] = x['model'].half()  # FP16 yap

    for p in x['model'].parameters():
        p.requires_grad = False

    save_path = save_path or path
    torch.save(x, save_path)

    mb = os.path.getsize(save_path) / 1E6
    print(f"Optimizer stripped from {path}, saved as {save_path}, size: {mb:.1f} MB")
````
